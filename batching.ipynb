```python
import json
```

```python
with open("../files/patterns_batch.json") as file:
    data = json.load(file)
print(data)
```

```python
tuningData = []
systemMessage = '''You are a trained data classifier that needs to classify a transaction based on certain parameters. You will receive a narration pattern that contains several tokens seperated by hyphens or whitespaces. Values enclosed in brackets are to be considered to be a single token. You are to analyse these tokens and their relative position in the pattern, and bin the pattern into a category and sub-category. You are to return only a single JSON dictionary for each pattern, with the keys 'type', 'category' and 'sub-category' with no additional text or information. The type denotes whether the transaction is credit or debit.'''
for i in data:
    typ, category, subcategory = i["type"], i["categroy"], i["sub-category"]
    diction = {"messages": [{"role": "system", "content": systemMessage}, {"role": "user", "content": i["pattern"]}, {"role": "assistant", "content": "{"+f'"type":"{typ}", "category":"{category}", "sub-category":"{subcategory}"'+"}"}]} #jsonl is very specific with the types of quotes used, only double quotes work
    tuningData.append(diction)
print(tuningData)
```

```python
trainingData, testingData = tuningData[0:int(0.7*len(tuningData))], tuningData[int(0.7*len(tuningData)):]
with open("../files/trainingData.jsonl", "w") as file:
    for data in trainingData:
        json.dump(data, file)
        file.write("\n")    #We need to add a newline for each entry for the jsonl format
with open("../files/testingData.jsonl", "w") as file:
    for data in testingData:
        json.dump(data, file)
        file.write("\n")
```
