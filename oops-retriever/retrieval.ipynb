```python
import pandas as pd
import random
import os
import json
```

```python
df = pd.read_csv("../files/Copy of cleaned_oops(1).csv")
```

```python
df
```

```python
df = df[(df['Part1 Total'] >= 20) & (df['Part1 Total']<30)]
```

```python
df
```

```python
random_sample = df.sample(n=17, random_state=20)
random_sample[['Part1 Total', 'Folder Name']]
```

```python
basePath = '../files/solutions_cleaned/'
selectedPath = '../files/selected_solutions/'
if not os.path.exists(selectedPath):
    os.mkdir(selectedPath)

infoDict = {}

for index, row in random_sample.iterrows():
    marks = row["Part1 Total"]
    folder = str(row['Folder Name'])
    infoDict[folder] = {}
    for i in range(1, 8):
        infoDict[folder]['Q'+str(i)] = row['Q'+str(i)]
    print(marks, folder)
    with open(basePath+folder+"/p1.java", "r") as file:
        code = file.read()
    with open(selectedPath+folder+".java", "w") as file:
        file.write(code)
    infoDict[folder]["Total"] = marks

with open(selectedPath+"info.json", "w") as file:
    json.dump(infoDict, file)
    
    
```

```python
with open(selectedPath+"out.csv", "w") as file:
    head = 'Solution, '
    for i in range(1, 8):
        head+='Q'+str(i)+', '
    head += 'Total, \n'
    file.write(head)
    for key in infoDict:
        writeStr = str(key)+', '
        for i in range(1, 8):
            writeStr += str(infoDict[key]['Q'+str(i)])+', '
        writeStr += str(infoDict[key]["Total"])
        file.write(writeStr)
        file.write('\n')
```
